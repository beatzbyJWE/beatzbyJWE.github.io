# Extracting and Comparing Year Transit Facility Rebuilt From Notes

## Loading in Data

In this exercise I will need the regex (re) package, pandas, and natural language toolkit. 

```python
import re
import pandas as pd
import nltk

pd.set_option('display.max_rows',200)
pd.set_option('display.max_colwidth',-1)
```

I am pulling in a published dataset that my team would like to work with. Each record holds data on a single transit agency facility, reported as of 2019.

```python
response='https://www.transit.dot.gov/sites/fta.dot.gov/files/2020-10/2019%20Facility%20Inventory.xlsx'
response
df=pd.read_excel(response)
```

There is a column for Year Built, as well as the date of the last condition assessment of the given facility, and any notes that the reporting transit agency felt were relevant to include. I only care about these particular records (along with the name of the agency) and I only care about records that have notes provided.

```python
al=df[['Agency Name','Year Built','Condition Assessment','Condition Assessment Date','Notes']]
Notes_only=al[al.Notes.notnull()] #filtering out all non-null notes
Notes_only
```

## Using String Contains & Pattern to Extract Year

Now I will use str.contains to find a four digit year, which can be in either the 1900s or the 2000s. The first command below will limit the Notes only table to Notes_year, a new dataframe with notes containing at least one four digit value.

```python
Notes_year=Notes_only[Notes_only['Notes'].str.contains('1[0-9]{3}','20[0-9]{2}')] #497 have rows that are probably years
Notes_year['Notes']
```

```python
with pd.option_context('display.max_colwidth', -1): display(Notes_year) #https://stackoverflow.com/questions/25351968/how-to-display-full-non-truncated-dataframe-information-in-html-when-convertin
```

## Extracting Condition Assessment from Notes

Now I will try to extract condition assessment, a single value, from the Notes section.

```python
pattern2='Condition'
Notes_CA=Notes_only[Notes_only['Notes'].str.contains('assessment')] #497 have rows that are probably years
Notes_CA
```

## Performing the (Imperfect) Comparison

Scanning through some of the notes above, it looks like the reported values in Year Built are very accurate compared to the notes. At this point, it isn't clear whether it is valuable to actually extract the year from the notes column to perform a logical comparison.

```python
pattern = '([0-9]{4})'

Notes_year['result'] = Notes_year.loc[:,'Notes'].str.extract(pattern)
Notes_year['result']=pd.to_numeric(Notes_year['result'])
Notes_year
```

So I can use an equivalence operator to add a new column, 'agree', which is a true/false value indicating whether the years are the same across the two fields.

```python
Notes_year['agree']=(Notes_year['result']==Notes_year['Year Built'])
Notes_year.value_counts('agree')
```

I can see from the above that there are only 39 results that agree. This seems like too many based on what I saw before. Upon further investigation I find out there must be a constraint in the system limiting the Year Built to 1900 and afterwards. So I adjust for that below.

```python
Notes_year[Notes_year['result']>1900].value_counts('agree')
```

## Conclusion

Now we see a much more balanced dataset. Let's check this set to make sure this is worthwhile.

```python
(Notes_year[Notes_year['result']>1900][Notes_year['result']<2021][Notes_year['agree']==False])
```

There are many values in this set that seem worth pursuing. Some are false positives as there were multiple values in the notes and the second one was right, where my query only captured the first one. 

What I would like to do in the future is go back and select the highest first, since "Year Built" is also defined as "Year Remodeled as New". In other words, the value captured from the notes may be describing the original year built, which would not be reported.
